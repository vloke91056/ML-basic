```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1,1,10)
```

```python
Y = X**2 + X + 1
```

```python
Y
```

```python
plt.plot(X,Y)
```

```python
#gradient descent optimization
x = 1
lr = 0.1
# gd = 2*x+1
plt.plot(X,Y)
for i in range(50):
  gd = 2*x+1
  x = x-lr*gd
  y = x**2 + x + 1
  print(x)
  plt.scatter(x,y)
plt.show()
```
